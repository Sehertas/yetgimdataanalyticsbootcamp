Pandas ve Numpy kütüphanelerini import ettik ve weather_data.csv dosyasını yükledik.

```python
import pandas as pd
import numpy as np
# CSV dosyasını okuyoruz
df = pd.read_csv('weather_data.csv')
```

# **3. SUTÜN SEÇİMİ**

Sutün seçtik ve buna göre tablomuzu analiz ettik.

```python
# Date,City ve Temperature sutünları seçelim.
df[['Date','City','Temperature']]
```

```python
#City ve temperature sutünlarını yanyana getirelim.
city_temp=df[['City','Temperature']]
print(city_temp)
```

# **4.FİLTRELEME**

Sutünları seçip, seçtiğimiz sutünları belirlediğimiz kurallara göre düzenledik.

```python
#Hava sıcaklığı(temperature) 30°C üzerinde olanları ve şehir olarak Bursa olanları sıralayalım.
#AND OPERATÖRÜ "&" İLE YAPILIR.
df[(df['Temperature'] > 30) & (df['City'] == 'Bursa')]
```

# **5**.**MANTIKSAL OPERATÖRLER VE** **FİLTRELEME**


Filtreleme işlemleri gerçekleştirdik. Birden fazla şartlı koşul yazdık.

```python
#Şehri(City) İstanbul ve Nem Oranı(Humidity) 60'tan büyük olanları sıralayalım.
#&=VE İŞLEMİ
df[(df['City']=='İstanbul') & (df['Humidity']>60)]
```

```python
#Şehri(City) Ankara olan veya sıcaklığı 5°C 'den küçük olanları sırala.
#or-> |
df[(df['City']=='Ankara') | (df['Temperature']>5)]
```

```python
#Sıcaklığı 10°C altında veya nemi>%70 üzerinde olan verileri filtreyelim.

df[(df['Temperature']<10) | (df['Humidity']>%70)]
```

# **6.SIRALAMA**

Verileri farklı sütunlara göre sıraladık.

```python
#Sıcaklık(Temperature) değerini artandan azalana sıralayalım. Ve ilk 10 kaydı gösterelim.
df.sort_values('Temperature', ascending=False)
df.head(10)
```

```python
#Nem değerlerini artandan azalana göre sıralayalım.
df.sort_values('Humidity', ascending=False)
```

```python
#Veriyi şehire göre artan olacak şekilde sıralayalım.
df.sort_values('City', ascending=True)
```

## **7. YENİ SUTÜN EKLEME**

Öncelikli olarak bir sutün ekledik. Fahrenheit cinsinden değer taşıması için matematiksel işlemler gerçekleştirdik.

```python
df['Temperature_F']=(df['Temperature']*9/5)+32
```

```python
df['FeelsLike']=(df['Temperature']-(df['Humidity'] / 100))
```
